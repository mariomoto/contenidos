```python
import requests
import textwrap

API_KEY = "REEMPLAZA_CON_TU_USER_KEY"
BASE_URL = "https://www.omdbapi.com/"

params = {
    "apikey": API_KEY,
    "t": "Inception",
    "plot": "full",
    "r": "json",
}

respuesta = requests.get(BASE_URL, params=params, timeout=10)
print("Status:", respuesta.status_code)

data = respuesta.json()
if data.get("Response") == "True":
    print("Titulo:", data.get("Title"))
    print("Ano:", data.get("Year"))
    print("IMDb:", data.get("imdbRating"))
    print("Plot:")
    print(textwrap.fill(data.get("Plot", ""), width=72))
else:
    print("Error OMDb:", data.get("Error", "Respuesta sin detalle de error"))
```

```python
import requests

COORDINADOR_API_KEY = "REEMPLAZA_CON_TU_USER_KEY"
BASE_COORDINADOR = "https://operacion.api.coordinador.cl"
endpoint = "/reportes/v3/generation"

params = {
    "user_key": COORDINADOR_API_KEY,
    "date": "2025-06-15",  # formato YYYY-MM-DD
}

respuesta = requests.get(BASE_COORDINADOR + endpoint, params=params, timeout=15)
print("Status:", respuesta.status_code)

if respuesta.status_code == 200:
    data = respuesta.json()
    print("Total de categorias:", data.get("totalElements"))
    for item in data.get("content", []):
        fuente = item.get("description")
        generacion = item.get("dailyCurrent")
        print(f"{fuente}: {generacion} GWh")
else:
    print("Error:", respuesta.text[:300])
```
